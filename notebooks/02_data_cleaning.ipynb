# Data Cleaning

Este notebook tem como objetivo realizar a limpeza e preparação dos dados utilizados no projeto.

Nesta etapa, serão tratados valores nulos, verificadas inconsistências e selecionadas as colunas mais relevantes para a análise do desempenho das equipes jogando em casa e fora de casa.

```python
import pandas as pd
```

```python
games = pd.read_csv('../data/raw/games.csv')
teams = pd.read_csv('../data/raw/teams.csv')
```

# Verificando Valores Nulos

Para mostrar quais colunas têm valores faltando.

```python
games.isnull().sum()
```

```python
teams.isnull().sum()
```

# Selecionando apenas as colunas importantes

```python
games_clean = games[['GAME_DATE_EST', 'HOME_TEAM_ID', 'VISITOR_TEAM_ID', 'SEASON', 'PTS_home', 'PTS_away', 'HOME_TEAM_WINS']].copy()
teams_clean = teams[['TEAM_ID', 'ABBREVIATION', 'NICKNAME', 'CITY']].copy()
```

# Verificando novamente se há valores nulos nas colunas escolhidas

```python
games_clean.isnull().sum()
```

```python
teams_clean.isnull().sum()
```

# Removendo linhas com dados ausentes relevantes

```python
games_clean = games_clean.dropna()
teams_clean = teams_clean.dropna()
```

## Verificação de duplicatas

A verificação de duplicatas será feita com cautela, pois em datasets esportivos pode haver registros que parecem repetidos, mas representam informações relevantes.

No dataset de jogos, a remoção de duplicatas só será realizada caso seja confirmado que se trata da mesma partida repetida sem diferenças relevantes.

```python
games_clean.duplicated().sum()
```

```python
teams_clean.duplicated().sum()
```

```python
games_clean[games_clean.duplicated(keep=False)].sort_values(by=['HOME_TEAM_ID', 'VISITOR_TEAM_ID', 'GAME_DATE_EST'])
```

```python
games_clean = games_clean.drop_duplicates()
```

# Verificando tipos de dados

```python
games_clean.info()
```

```python
teams_clean.info()
```

# Convertendo data

```python
games_clean['GAME_DATE_EST'] = pd.to_datetime(games_clean['GAME_DATE_EST'])
```

# Visualizando resultado da limpeza

```python
games_clean.head()
```

```python
teams_clean.head()
```

# Salvando os dados tratados

```python
games_clean.to_csv('../data/processed/clean_games.csv', index=False)
teams_clean.to_csv('../data/processed/clean_teams.csv', index=False)
```

## Resultado da limpeza

Foram selecionadas apenas as colunas mais relevantes para a análise do projeto.

Também foram tratados valores nulos, removidas possíveis duplicatas e ajustados os tipos de dados, deixando a base pronta para a próxima etapa: análise exploratória.
