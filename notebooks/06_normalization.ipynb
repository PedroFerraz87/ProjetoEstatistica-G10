```python
import pandas as pd

df = pd.read_csv('../data/processed/clean_games.csv')

df.head()
```

```python
df['GAME_DATE_EST'] = pd.to_datetime(df['GAME_DATE_EST'])
df['date_numeric'] = df['GAME_DATE_EST'].map(pd.Timestamp.toordinal)
```

```python

df['pts_home_z'] = (df['PTS_home'] - df['PTS_home'].mean()) / df['PTS_home'].std()
df['pts_away_z'] = (df['PTS_away'] - df['PTS_away'].mean()) / df['PTS_away'].std()
df['date_z'] = (df['date_numeric'] - df['date_numeric'].mean()) / df['date_numeric'].std()
```

```python
df.to_csv('../data/processed/games_normalized.csv', index=False)
```

## Normalização dos Dados

Foi aplicada a técnica de padronização utilizando o método Z-score nas variáveis de pontuação (casa e visitante) e na variável de data.

Essa transformação foi necessária pois as variáveis possuem escalas diferentes. A variável de data apresenta valores muito maiores que as pontuações, o que poderia distorcer análises futuras.

O Z-score padroniza os dados para média 0 e desvio padrão 1, permitindo comparações justas entre as variáveis.
