## Médias de pontuação

Aqui serão analisadas as médias de pontos marcados pelas equipes mandantes e visitantes, permitindo uma comparação inicial entre desempenho em casa e fora.

```python
print(f"Média de pontos do time da casa: {games['PTS_home'].mean():.2f}")
print(f"Média de pontos do time visitante: {games['PTS_away'].mean():.2f}")
```

## Medidas de Centralização — Moda

Além da média, a **moda** é uma medida de centralização importante: ela indica o valor que aparece com maior frequência na distribuição. Para variáveis contínuas como a pontuação, a moda revela qual placar é mais recorrente nas partidas.

```python
from scipy import stats

moda_home = games['PTS_home'].mode()[0]
moda_away = games['PTS_away'].mode()[0]
mediana_home = games['PTS_home'].median()
mediana_away = games['PTS_away'].median()

print('=== Medidas de Centralização — Pontuação ===')
print(f'Média   do time da casa:   {games["PTS_home"].mean():.2f}')
print(f'Mediana do time da casa:   {mediana_home:.2f}')
print(f'Moda    do time da casa:   {moda_home}')
print()
print(f'Média   do time visitante: {games["PTS_away"].mean():.2f}')
print(f'Mediana do time visitante: {mediana_away:.2f}')
print(f'Moda    do time visitante: {moda_away}')
```

A **moda** dos pontos dos times da casa e visitantes revela o placar mais comum nas partidas. Quando a moda se aproxima da mediana e da média, a distribuição é aproximadamente simétrica. Diferenças entre esses três valores indicam assimetria na distribuição de pontuações.

## Estatísticas descritivas das pontuações

A tabela acima apresenta estatísticas descritivas das colunas `PTS_home` e `PTS_away`, que representam, respectivamente, os pontos marcados pelo time da casa e pelo time visitante.

A linha **count** indica a quantidade de registros válidos disponíveis em cada coluna, mostrando que a base possui 26.523 jogos com pontuação registrada para ambos os casos.

A linha **mean** representa a média de pontos. Observa-se que os times da casa tiveram média de **103,45 pontos**, enquanto os visitantes tiveram média de **100,63 pontos**, sugerindo um desempenho ofensivo ligeiramente superior das equipes mandantes.

A linha **std** corresponde ao desvio padrão, que mostra o quanto os valores costumam variar em relação à média. Nesse caso, tanto os pontos dos times da casa quanto dos visitantes apresentam variação semelhante, em torno de 13 pontos.

A linha **min** mostra a menor pontuação registrada, enquanto a linha **max** indica a maior pontuação observada em cada coluna.

As linhas **25%**, **50%** e **75%** representam os quartis da distribuição. O valor de **50%** corresponde à mediana, isto é, o ponto central dos dados. Nota-se que a mediana de `PTS_home` é **103**, enquanto a de `PTS_away` é **100**, reforçando a tendência de maior pontuação das equipes jogando em casa.

De modo geral, essa tabela indica que os times mandantes apresentam, em média, uma pontuação um pouco superior à dos visitantes, o que pode ser um primeiro indício de influência do fator casa no desempenho das equipes.

## Medidas de Dispersão — Variância, Desvio Padrão e Amplitude

As medidas de dispersão quantificam o grau de espalhamento dos dados em torno da média. Além do desvio padrão (já presente no `.describe()`), a **variância** e a **amplitude** completam o conjunto mínimo exigido de medidas de dispersão.

```python
print('=== Medidas de Dispersão — Pontuação ===')
for col, label in [('PTS_home', 'Time da Casa'), ('PTS_away', 'Time Visitante')]:
    variancia  = games[col].var()
    desvio_pad = games[col].std()
    amplitude  = games[col].max() - games[col].min()
    print(f'\n{label}:')
    print(f'  Variância:      {variancia:.2f}')
    print(f'  Desvio Padrão:  {desvio_pad:.2f}')
    print(f'  Amplitude:      {amplitude} (min={games[col].min()}, max={games[col].max()})')
```

A **variância** mede a dispersão quadrática dos valores em relação à média — quanto maior, mais espalhados estão os pontos. O **desvio padrão** é a raiz quadrada da variância e tem a mesma unidade da variável original (pontos), sendo mais intuitivo para interpretação. A **amplitude** indica a diferença entre o maior e o menor placar registrado na base.
